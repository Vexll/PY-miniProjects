```python
import requests
import pandas as pd
from datetime import date
```

```python
def check_format(num: int) -> str:
    if num < 10:
        return f"0{num}"
    else:
        return f"{num}"

def get_previous_date(t_date: str) -> str:
    splitting_date = t_date.split('-')
    year = int(splitting_date[0])
    month = int(splitting_date[1])
    day = int(splitting_date[2])
    
    # checks if the day = 1
    if day == 1:
        # checks if the month = 1
        if month == 1:
            year = year - 1
            month = 12
        else:
            month = month - 1    
        day = 30
    else:
        day = day - 1 
    
    return f"{year}-{check_format(month)}-{check_format(day)}"
    
        
```

```python
PRICING_API_KEY = ""
STOCK_NAME = "TSLA"
```

```python
params = {
    'apikey': PRICING_API_KEY,
    'symbol': STOCK_NAME,
    'function': 'TIME_SERIES_DAILY',
}
```

```python
url = "https://www.alphavantage.co/query"
response = requests.get(url, params=params)
```

```python
response
```

```python
data = response.json()
```

```python
data
```

```python
def get_percentage_change(t_p, p_p) -> str:
    percentage = ((t_p - p_p) / p_p) * 100
    return f"{percentage:.2f}%"
```

```python
today_date = "2024-05-15"
today_price = data["Time Series (Daily)"][today_date]['4. close']
prev_date = get_previous_date(today_date)
prev_price = data["Time Series (Daily)"][prev_date]['4. close']
```

```python
print(f"stock price of {today_date}:\t{today_price}\nstock price of {prev_date}:\t{prev_price}")
```

```python
print(f"percentage change = {get_percentage_change(float(today_price), float(prev_price))}")
```
